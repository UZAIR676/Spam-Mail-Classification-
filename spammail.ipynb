```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer #to convert the text into number
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score
```

```python
raw_mail_data = pd.read_csv("/content/mail_data.csv")
```

```python
raw_mail_data
```

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

```python
mail_data.shape
```

```python
mail_data.describe() # Changed 'discribe' to 'describe'
```

```python
# now we will covert the hmm to 1 and spam to 0
mail_data.loc[mail_data['Category']=='spam','Category', ]=0
mail_data.loc[mail_data['Category']=='ham','Category', ]=1
```

```python
mail_data.head()
```

```python
x = mail_data['Message']
y = mail_data['Category']
```

```python
x
```

```python
y
```

splitting tht training data and disk data


```python
x_train ,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype("int")
```

```python
print(x_train_features)
```

```python
model = LogisticRegression()
```

```python
# training the logistic regression model
model.fit(x_train_features,y_train)
```

```python
# evaluating the train model
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data :',accuracy_on_training_data)
```

```python
# evaluating the test model
prediction_on_testing_data = model.predict(x_test_features)
accuracy_on_testing_data = accuracy_score(y_test,prediction_on_testing_data)
```

```python
print("Accuracy on the training data :",accuracy_on_testing_data)
```
